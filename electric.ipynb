```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\Ganesh\Downloads\Electric_Vehicle_Population_Data (1).csv"
try:
    data = pd.read_csv(file_path)
    print("Dataset imported successfully!")
except Exception as e:
    print(f"Error importing dataset: {e}")

# Display basic info about the dataset
print("\nDataset Information:")
print(data.info())

# Display first few rows of the dataset
print("\nFirst 5 rows:")
print(data.head())

# Data cleaning: Handle missing values
print("\nHandling missing values...")
data.ffill(inplace=True)  # Forward-fill missing values
print("Missing values handled.")

# Data manipulation: Add a new column using 'Electric Range'
print("\nAdding a new column based on 'Electric Range'...")
if 'Electric Range' in data.columns:
    data['New_Column'] = data['Electric Range'] * 2
    print("New column added.")
else:
    print("Column 'Electric Range' not found in the dataset!")

# Plotting a graph using 'Model Year' and 'Electric Range'
print("\nPlotting a graph...")
if 'Model Year' in data.columns and 'Electric Range' in data.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(data['Model Year'], data['Electric Range'], color='blue', alpha=0.5)
    plt.xlabel('Model Year')
    plt.ylabel('Electric Range')
    plt.title('Electric Range vs Model Year')
    plt.grid(True)
    plt.show()
else:
    print("Required columns for plotting are not found!")

# Save cleaned and manipulated data
output_file_path = r"C:\Users\Ganesh\Downloads\cleaned_dataset.csv"
data.to_csv(output_file_path, index=False)
print(f"Cleaned and manipulated dataset saved to: {output_file_path}")
```
